## V Selecionando Apenas um tipo (Imoveis Residenciais)

```python
# Importa a base denovo

dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# Cria uma lista com os valores sem duplicatas

list(dados['Tipo'].drop_duplicates())
```

```python
# Seleciona e guarda os valores que interessa em outra lista

residencial = ['Quitinete',
'Casa',
 'Apartamento',
 'Casa de Condomínio', 
 'Casa de Vila', 
 'Chácara',
 'Loteamento/Condomínio',
 'Sítio',
 ]

residencial
```

```python
# Cria uma Series booleana retornando true na coluna onde for encontrado algum valor que estivesse contido na lista

selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# Cria um DF com apenas as linhas True da seleçao

dadosResidenciais = dados[selecao]
dadosResidenciais
```

```python
#Verifica os tipos unicos para conferir

list(dadosResidenciais['Tipo'].drop_duplicates())
```

```python
# Reconstroi o index

dadosResidenciais.index = range(dadosResidenciais.shape[0])
dadosResidenciais
```

## VI Exportando a Base de Dados

## VII Seleções e Frequencia + Export (Revisão)

```python
# Importando a base alterada nas seções anteriores

dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/aluguelResidencial.csv', sep = ';')
dados
```

```python
# Selecione somente os imoveis classificados com o tipo "Apartamento"

apt = ['Apartamento']
selecao = dados['Tipo'].isin(apt)
apts = dados[selecao]
apts.index = range(apts.shape[0])
apts
```

```python
# Selecione os imoveis classificados com tipo 'Casa', 'Casa de Condominio' e 'Casa de Vila'

casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
selecao = dados['Tipo'].isin(casa)
casas = dados[selecao]
casas.index = range(casas.shape[0])
casas
```

```python
# Selecione os imoveis com área entre 60 e 100 metros quadrados, incluindo os limites

selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100)
areas = dados[selecao]
areas.index = range(areas.shape[0])
areas
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
imoveisQuaQua = dados[selecao]
imoveisQuaQua.index = range(imoveisQuaQua.shape[0])
imoveisQuaQua
```

```python
#Exportando as bases

apts.to_csv('/content/drive/MyDrive/Colab Notebooks/data/apartamentosParaAluguel.csv', sep =';', index = False)
casas.to_csv('/content/drive/MyDrive/Colab Notebooks/data/casasParaAluguel.csv', sep =';', index = False)
imoveisQuaQua.to_csv('/content/drive/MyDrive/Colab Notebooks/data/quatroQuartos2000$ParaAluguel.csv', sep =';', index = False)
areas.to_csv('/content/drive/MyDrive/Colab Notebooks/data/imoveis60-100mParaAluguel.csv', sep =';', index = False)
```
